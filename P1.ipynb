```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2, 2.1, 0.1)

y_sin = np.sin(x)

x_positive = x[x > 0]
y_log = np.log10(x_positive)

y_exp = np.exp(x)


print("y = sin(x) = ", y_sin)
print("y = log10(x) = ", y_log)
print("y = e^x = ", y_exp)

plt.subplot(2, 2, 1)
plt.plot(x, y_sin)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sin(x)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x_positive, y_log)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = log10(x)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x, y_exp)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = e^x')
plt.grid(True)
plt.show()
```

1.9. Realizar una funcin que despliegue los siguientes patrones utilizando asteriscos.

```python
base = 10
altura = 10

for i in range(1, altura + 1):
    #calcula cantidad de espacios y asteriscos del 1er triangulo
    astTriangulo1 = i * "*"
    espTriangulo1 = " " * (base - i)
    #calcula cantidad de espacios y asteriscos del 2do triangulo
    astTriangulo2 = "*" * (base + 1 - i)
    espTriangulo2 = " " * (i - 1)

    espDefault = " " * 5

    #concatena 1er y 2do  triangulo,junto con espacios
    triangulos1 = astTriangulo1 + espTriangulo1 + espDefault + astTriangulo2 + espTriangulo2
    #invierte el resultado anterior
    triangulos2 = triangulos1[::-1]

    print(triangulos1 + espDefault + triangulos2)
```

1.10. Realizar una funcin que despliegue un tringulo utilizando asteriscos. La altura ser definida por el
usuario, para el ejemplo de la figura la altura es 5.

```python
altura = int(input("ingresa la altura del triangulo: "))

for i in range(1, altura + 1):
        espacios = altura - i
        asteriscos = 2 * i - 1
        linea = " " * espacios + "*" * asteriscos
        print(linea)
```
